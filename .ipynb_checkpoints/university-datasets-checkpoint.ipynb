```python
# Lets take a look at the missing data
missing_data = uni[uni.isin(['-']).any(axis=1)]
missing_data
```

```python
# The missing data is denoted by the character '-', below we can construct a data frame using only complete record
x = ['-']
uni_complete = uni[ (uni.isin(x).any(axis=1)==False )] 
uni_complete
```
